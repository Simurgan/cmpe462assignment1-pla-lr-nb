```python
from pla import PLA
```

```python
small_pla = PLA('PLA_data/data_small.npy', 'PLA_data/label_small.npy')
large_pla = PLA('PLA_data/data_large.npy', 'PLA_data/label_large.npy')
```

```python
# Q1.1
print("----------------------------------------------------------")
small_pla.train()
print("PLA training stats for small dataset:")
small_pla.print_last_training_stats()
print("----------------------------------------------------------")
large_pla.train()
print("PLA training stats for large dataset:")
large_pla.print_last_training_stats()
print("----------------------------------------------------------")
```

```python
# Q1.2.1
small_pla.scatter_data()
small_pla.plot_decision_boundary()
```

```python
# Q1.2.2
large_pla.scatter_data()
large_pla.plot_decision_boundary()
```

```python
# Q1.3
import random

for i in range(5):
    print("----------------------------------------------------------")
    print(f"Training PLA with random initial weights #{i + 1}")
    initial_weights = [random.uniform(-1 * 10**(i + 1), 10**(i + 1)) for k in range(3)]
    print("Initial weights:", initial_weights)
    small_pla.train(initial_weights)
    small_pla.print_last_training_stats()
    small_pla.plot_decision_boundary()

print("----------------------------------------------------------")
```
